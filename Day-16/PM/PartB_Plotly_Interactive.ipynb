# Day 16 PM – Part B: Plotly Express Interactive Charts

**Recreating 3 charts from Part A as interactive Plotly**

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

df = pd.read_csv('eda_assignment_data.csv')
print(df.head())
```

## 1. Interactive Histogram (Total Spent)


```python
fig1 = px.histogram(df, x='total_spent', nbins=50, 
                   title='Interactive: Total Spent Distribution',
                   labels={'total_spent': 'Total Spent (₹)'},
                   marginal='rug')
fig1.update_layout(height=500)
fig1.write_html('plotly_histogram_total_spent.html')
fig1.show()
```

## 2. Interactive Scatter + Trendline (Orders vs Total Spent)


```python
fig2 = px.scatter(df, x='orders_last_month', y='total_spent', 
                 trendline='ols',
                 title='Interactive: Orders vs Total Spent',
                 labels={'orders_last_month': 'Orders Last Month', 'total_spent': 'Total Spent (₹)'},
                 hover_data=['age', 'segment'])
fig2.update_layout(height=500)
fig2.write_html('plotly_scatter_orders_spent.html')
fig2.show()
```

## 3. Interactive Box Plot (Total Spent by Segment)


```python
fig3 = px.box(df, x='segment', y='total_spent', 
             title='Interactive: Total Spent by Segment',
             labels={'total_spent': 'Total Spent (₹)'},
             points='outliers+ suspectedoutliers')
fig3.update_layout(height=500)
fig3.write_html('plotly_box_segment.html')
fig3.show()
```

## Interactive vs Static Insights

**1. Hover details**: See exact `age`/`segment` for outlier points.
**2. Zoom/pan**: Explore dense regions of scatter plot (e.g., orders=0-5).
