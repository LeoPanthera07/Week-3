## Q1: Why is this slow + vectorized fix

**Problem**: Double Python loop → **O(n²)** interpreter calls, no NumPy speed.

**Speedup**: ~1000x for 1000×1000 (NumPy C-optimized vs Python loops).

**Vectorized** (1 line):
```python
result = data**2 + 2*data + 1  # Broadcasting/element-wise
```

````python
def k_nearest(data: np.ndarray, point: np.ndarray, k: int) -> np.ndarray:
    """Vectorized kNN - returns indices of k closest points."""
    diffs = data - point
    dists = np.sum(diffs**2, axis=1)**0.5
    return np.argsort(dists)[:k]

# Test
data = np.random.rand(100, 2)
point = np.array([0.5, 0.5])
nearest = k_nearest(data, point, 5)
print('5 nearest indices:', nearest)
````

## Q3: Debug column normalization

**3 Bugs**:
1. **`axis=1`** → **row** means/std (should be `axis=0` for columns)
2. **Shape mismatch**: `means` shape `(100,)` cannot broadcast to `(100,5)`
3. **No centering**: Subtracts row-mean from entire row → meaningless

**Fixed**:
```python
means = data.mean(axis=0, keepdims=True)  # (1,5)
stds  = data.std(axis=0, keepdims=True)   # (1,5)
normalized = (data - means) / stds
```
