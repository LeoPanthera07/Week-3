```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(1313)

temp     = rng.uniform(15, 45, (50, 24))
humidity = rng.uniform(20, 95, (50, 24))
battery  = rng.uniform(10, 100, (50, 24))

data = np.stack([temp, humidity, battery], axis=2)
print('Shape:', data.shape)
```

```python
# Task 2: Alert sensors (temp>40 OR humidity>90 in ANY hour)
alert_mask = np.any((data[:, :, 0] > 40) | (data[:, :, 1] > 90), axis=1)
alert_sensors = np.where(alert_mask)[0]
print(f'Alert sensors: {alert_sensors}')
```

```python
# Task 3: Per-sensor daily averages (50x3)
daily_avg = data.mean(axis=1)
print('Daily averages shape:', daily_avg.shape)
print('Sample sensor 0:', daily_avg[0].round(2))
```

```python
# Task 4: Hour with highest average temperature across all sensors
hourly_temp_avg = data[:, :, 0].mean(axis=0)
hottest_hour = np.argmax(hourly_temp_avg)
print(f'Hottest hour: {hottest_hour} ({hottest_hour}:00)')
```

```python
# Task 5: Battery drain (first - last hour), flag >50% drain
battery_drain = data[:, 0, 2] - data[:, -1, 2]
critical_drain = np.where(battery_drain > 50)[0]
print(f'Sensors with critical battery drain (>50%): {critical_drain}')
```

```python
# Task 6: Min-max normalize all metrics per metric (axis over sensors+hours)
min_vals = data.reshape(-1, 3).min(axis=0)
max_vals = data.reshape(-1, 3).max(axis=0)
normalized = (data - min_vals) / (max_vals - min_vals)
print('Normalized range check:', normalized.min().round(4), normalized.max().round(4))
```

```python
# Task 7: Save daily averages as sensor_summary.csv
df_summary = pd.DataFrame(daily_avg, columns=['avg_temperature', 'avg_humidity', 'avg_battery'])
df_summary.to_csv('sensor_summary.csv', index=False)
print('sensor_summary.csv saved — shape:', df_summary.shape)
print(df_summary.head())
```
