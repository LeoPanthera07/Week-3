```python
import numpy as np

# 1. 3x3 matrix analysis
rng = np.random.default_rng(42)
A = rng.normal(0, 2, (3, 3))
print('Matrix A:\n', A.round(2))
```

```python
# Determinant, inverse, eigenvalues
det_A = np.linalg.det(A)
A_inv = np.linalg.inv(A)
eigenvals, eigenvecs = np.linalg.eig(A)

print('Determinant:', round(det_A, 4))
print('A @ A_inv ≈ I:', np.allclose(A @ A_inv, np.eye(3)))
print('Eigenvalues:', np.round(eigenvals, 2))
```

```python
# 2. Solve linear system: 2x + 3y = 8, 4x + y = 10
A_sys = np.array([[2, 3], [4, 1]])
b_sys = np.array([8, 10])
solution = np.linalg.solve(A_sys, b_sys)
print('Solution x, y:', solution)
print('Verification:', np.allclose(A_sys @ solution, b_sys))
```

## 3. np.linalg.svd() in ML

**SVD** decomposes matrix A = UΣVᵀ (U/V orthogonal, Σ diagonal singular values).

**ML uses**:
- **PCA**: Truncated SVD → dimensionality reduction
- **Recommendation systems**: Matrix factorization (user-item ratings)
- **Image compression**: Low-rank approximation
- **Pseudoinverse**: Solving under/over-determined systems
