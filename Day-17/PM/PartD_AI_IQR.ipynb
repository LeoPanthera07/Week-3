## 1. Exact Prompt
`Write a NumPy function that performs IQR-based outlier detection on each column of a 2D array, replacing outliers with the column median.`

## 2. AI Output

```python
def remove_outliers_iqr(arr):
    median = np.median(arr, axis=0)
    q1 = np.percentile(arr, 25, axis=0)
    q3 = np.percentile(arr, 75, axis=0)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    outliers = (arr < lower_bound) | (arr > upper_bound)
    arr[outliers] = median[outliers]
    return arr
```

````python
def ai_iqr_outliers(arr):
    median = np.median(arr, axis=0)
    q1 = np.percentile(arr, 25, axis=0)
    q3 = np.percentile(arr, 75, axis=0)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = (arr < lower) | (arr > upper)
    arr[outliers] = median[outliers]
    return arr

# Test with known outliers
test_data = np.random.normal(0, 1, (100, 3))
test_data[0, 0] = 4.5  # Outlier
print('Before:\n', test_data[[0,1,2], :])
print('After:\n', ai_iqr_outliers(test_data.copy())[[0,1,2], :])
````
