```python
import numpy as np
import time

# 1. Compare new vs legacy API
size = 1_000_000

# Legacy
t0 = time.time()
np.random.seed(42)
legacy_samples = np.random.normal(0, 1, size)
legacy_time = time.time() - t0

# New API
t1 = time.time()
rng = np.random.default_rng(42)
new_samples = rng.normal(0, 1, size)
new_time = time.time() - t1

print(f'Legacy: {legacy_time:.3f}s | New: {new_time:.3f}s')
print('Samples match:', np.allclose(legacy_samples, new_samples))
```

```python
# 2. Linear regression synthetic dataset generator
def synthetic_regression_dataset(n_samples=100, n_features=3, seed=42):
    rng = np.random.default_rng(seed)
    true_weights = np.array([2.5, -1.3, 0.7])
    
    X = rng.normal(0, 1, (n_samples, n_features))
    noise = rng.normal(0, 0.5, n_samples)
    y = X @ true_weights + noise
    
    return X, y, true_weights

X, y, w_true = synthetic_regression_dataset()
print('X shape:', X.shape)
print('True weights:', w_true)
print('w_est (OLS):', np.linalg.lstsq(X, y, rcond=None)[0].round(2))
```

## 3. np.random.Generator explanation

**What**: Modern BitGenerator-based PRNG replacing legacy Global RandomState.

**Why introduced**:
- **Thread-safe**: Separate RNG per thread, no race conditions.
- **Reproducible**: Same seed → same sequence across Python/NumPy versions.
- **Faster**: PCG64, Philox algorithms outperform legacy Mersenne Twister.
- **Flexible**: Multiple BitGenerators (PCG64, MT19937, etc.).

**Usage**: `rng = np.random.default_rng(seed)` → thread-local, reproducible, fast.
