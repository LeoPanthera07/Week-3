## Q1: Broadcasting Rules (Analogy + Formal + Examples)

**Real-world analogy**: Like a teacher asking the same question to 30 students — one answer applies to everyone (broadcasting smaller array to match larger one).

**3 Formal Rules**:
1. Dimensions aligned from trailing (right) edge
2. Size 1 dimension expands to match other
3. Dimensions must be **equal or 1** (no broadcasting if 3≠2)

````python
# WORKS: (3,) + (3,1) → both become (3,3)
a = np.array([1, 2, 3])
b = np.array([[10],[20],[30]])
print('Works:', (a[:, np.newaxis] + b).shape)
````

````python
# FAILS: (2,3) + (3,) → trailing dims 3≠2
try:
    c = np.random.rand(2,3)
    d = np.random.rand(3)
    print(c + d)
except ValueError as e:
    print('Fails:', e)
````

## Q2: row_normalize() function

Each row → probability distribution (sum=1). Zero-sum rows unchanged.

````python
def row_normalize(arr: np.ndarray) -> np.ndarray:
    """Normalize each row to sum to 1. Zero-sum rows stay zeros."""
    row_sums = arr.sum(axis=1, keepdims=True)
    return np.where(row_sums > 0, arr / row_sums, 0)

# Test
test_data = np.array([[1, 2, 3], [0, 0, 0], [4, 0, 2]])
print('Original:\n', test_data)
print('Normalized:\n', row_normalize(test_data))
````

## Q3: Debug the subtle bug

**BUG**: `data > 2 and data < 5` → **element-wise** `and` doesn't work on arrays.

**Issues**:
1. `and` tries scalar boolean → `ValueError`
2. `reshape(2,1)` assumes exactly 2 elements → fails for other masks

**Fixed**:
```python
data = np.array([1, 2, 3, 4, 5])
mask = (data > 2) & (data < 5)  # Bitwise &
filtered = data[mask]
result = filtered[:, np.newaxis]  # General reshape
print(result)  # [[3], [4]]
```
