## 1. Exact Prompt
`Write a NumPy function that performs min-max normalization on a 2D array, scaling each column to [0, 1] range.`

## 2. AI Output (Claude/GPT/Claude style)

```python
def min_max_normalize(arr):
    """
    Min-max normalize each column to [0,1].
    """  
    min_vals = np.min(arr, axis=0)
    max_vals = np.max(arr, axis=0)
    return (arr - min_vals) / (max_vals - min_vals)
```

````python
def ai_min_max_normalize(arr):
    min_vals = np.min(arr, axis=0)
    max_vals = np.max(arr, axis=0)
    return (arr - min_vals) / (max_vals - min_vals)

# Test
test_arr = np.array([[1, 100], [3, 200], [5, 300]])
print('Original:\n', test_arr)
print('AI Normalized:\n', ai_min_max_normalize(test_arr))
````
