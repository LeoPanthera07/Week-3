```python
import numpy as np

rng = np.random.default_rng(2026)
grades = rng.integers(0, 101, size=(200, 8))
print('Shape:', grades.shape)
print('Sample row 0:', grades[0])
```

```python
student_avg = grades.mean(axis=1)
course_avg  = grades.mean(axis=0)
overall_mean = grades.mean()
overall_std  = grades.std()

print('Per-course averages (before curve):', np.round(course_avg, 1))
print('Overall mean:', round(overall_mean, 2))
print('Overall std :', round(overall_std,  2))
```

```python
curved = grades.copy().astype(float)
needs_curve = course_avg < 50
print('Courses getting curve:', np.where(needs_curve)[0])

curved[:, needs_curve] = np.minimum(curved[:, needs_curve] + 10, 100)
print('Per-course averages (after curve):', np.round(curved.mean(axis=0), 1))
```

```python
letter_grades = np.where(curved >= 90, 'A',
               np.where(curved >= 80, 'B',
               np.where(curved >= 70, 'C',
               np.where(curved >= 60, 'D', 'F'))))

print('Letter grade shape:', letter_grades.shape)
print('Sample student 0 grades:', letter_grades[0])
```

```python
student_avg_curved = curved.mean(axis=1)
top10_indices = np.argsort(student_avg_curved)[::-1][:10]
print('Top 10 student indices:', top10_indices)
print('Their averages:', np.round(student_avg_curved[top10_indices], 1))
```

```python
passed_all = np.all(curved >= 60, axis=1)
print(f'Students passing all courses: {passed_all.sum()} out of 200')
```
